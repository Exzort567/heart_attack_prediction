SPLITTING THE FEATURES AND TARGET

```python
x = df.drop(columns = 'output', axis = 1)
y = df['output']
```

```python
print(x)
```

```python
print(y)
```

SPLITTING THE DATASET INTO TRAINING DATA AND TESTING

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify = y, random_state =2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train, x_test)
```

MODEL TRAINING


```python
accuracies = {}
from sklearn.linear_model import LogisticRegression
import pickle
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

TESTING

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score, r2_score
```

```python
acc = accuracy_score(y_test, y_pred)*100
f1 = f1_score(y_test, y_pred)*100
prec = precision_score(y_test, y_pred)*100
rec = recall_score(y_test, y_pred)*100
r22 = r2_score(y_test, y_pred)*100
cn = confusion_matrix(y_test, y_pred)
```

```python
result = pd.DataFrame([['LogisticRegression', acc, f1, prec, rec, r22]],
                      columns = ['Model', 'Accuracy', 'F1 Score', 'Precision Score', 'Recall', 'R2 Score'])

with open('logistic_regression_model.pkl', 'wb') as f:
    pickle.dump(lr, f)


print(result)
```
